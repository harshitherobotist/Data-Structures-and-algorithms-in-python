# LINKED LIST-

Advantages over array-

1)dynamic size(No need to guess the space required)\n

2)ease of insertion deletion (as we only have to change pointer address)

Disadvantages-

1)Linear Transversing Time

2)Extra memory for pointer allocation

```python
#Linked list is represented by a pointer to the head
class Node:
    
    def __init__(self,data):
        self.data = data  #assigning data
        self.next = None  #initializing next as null

class LinkedList:
    
    def __init__(self):
        self.head = None
    def printlist(self):
        temp = self.head
        while(temp):
            print(temp.data, end =" ")
            temp = temp.next
    def push(self,new_data):
        """to insert new node at beginning"""
        new_node = Node(new_data)
        
        new_node.next = self.head  #make next of new node as head
        
        self.head = new_node  #move the head of pointer to the new node
    
    def deleteNode(self,key):
        """delets the node at a particular key"""
        temp = self.head
        
        #if head contains the key
        if(temp is not None):
            if (temp.data == key):
                self.head = temp.next
                temp = None
                return
        #search for key to be deleted keep track of the  
        # previous node as we need to change 'prev.next
        while(temp is not None):
            if temp.data == key:
                break
            prev = temp
            temp = temp.next
        #if key not present in the list
        if (temp is None):
            return
        prev.next = temp.next
        temp =None
            
    
    def InsertAfter(self,prev_node,new_data):
        """to insert a node after a given node"""
        #check if node is present
        if prev_node is None:
            return print("The given node should be in the linked list")
        
        new_node = Node(new_data)  #initialize new node
        
        new_node.next = prev_node.next  #set pointer of new node to next node
        
        prev_node.next = new_node   #set pointer of prev node to new node
        
    def append(self,new_data):
        """to insert a node at the end
        This method can also be optimized to work in O(1) by keeping an extra pointer to tail of linked list"""
        new_node = Node(new_data)
        
        #if linked list is empty make the new node as head
        if self.head is None:
            self.head = new_node
            return
        #else traverse till the last node
        last = self.head
        while(last.next):
            last = last.next
        last.next = new_node
        
        
if __name__ == '__main__':
    
    #creating an empty list
    llist = LinkedList()
    
    llist.head = Node(1)
    second = Node(2)
    third = Node(3)
    #now there nodes are created but now lets create pointers
    llist.head.next = second
    second.next = third
    
    #printing the list
    llist.printlist()
    llist.append(4)
    llist.append(5)
    llist.push(1)
    llist.InsertAfter(second,5)
    llist.deleteNode(5)
    print("\n")
    llist.printlist()
        
        
    
```
